```python
import json
import pandas as pd
```

```python
with open("risultatiRegione.json", "r") as f:
    eligendo=json.load(f)
```

```python
d=[]
```

```python
for e in eligendo:
    c = e["comuneRes"]
    comune=c["int"]["desc_com"]
    print(comune)
    cod_com=c["int"]["cod_com"]
    cod_prov=c["int"]["cod_prov"]
    for cand in c["cand"]:
        res={}
        res["comune"]=comune
        res["cod_com"]=cod_com
        res["cod_prov"]=cod_prov
        res["cand_cogn"]=cand["cogn"]
        res["cand_nome"]=cand["nome"]
        res["cand_voti"]=cand["voti"]
        res["cand_perc"]=cand["perc"]
        d.append(res)
        
```

```python
df=pd.DataFrame(d)
```

```python
df.loc[df["comune"].str.contains("MONTESCUDO"),"comune"]="MONTESCUDO-MONTE COLOMBO"
df.loc[df["comune"].str.contains("FORLI'"),"comune"]="FORLÃŒ"
```

```python
df.to_csv("candidati_comune.csv")
```
